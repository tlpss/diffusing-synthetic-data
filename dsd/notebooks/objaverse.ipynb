```python
import objaverse
import open3d as o3d
```

```python
uids = objaverse.load_uids()
```

```python
anns  = objaverse.load_annotations(uids[:20000])
```

```python
mug_uids = [uid for uid, ann in anns.items() if 'mug' in [x["name"] for x in ann['tags']]]
```

```python
mug_uids
```

```python
mug_meshes = objaverse.load_objects(mug_uids)
```

```python
mug_meshes.values().__iter__().__next__()
```

```python
import trimesh

mesh = trimesh.load(list(mug_meshes.values().__iter__())[2])
```

```python
mesh.export("mug.obj")
```

```python
from dsd import DATA_DIR
target_dir = DATA_DIR / "meshes" /"objaverse mugs"

for uuid, mesh_path in mug_meshes.items():
    mesh = trimesh.load(mesh_path)
    mesh_dir = DATA_DIR / "meshes" / "objaverse mugs" / f"{uuid}/"
    mesh_dir.mkdir(exist_ok=True, parents=True)
    mesh.export( mesh_dir / f"{uuid}.obj")
```
