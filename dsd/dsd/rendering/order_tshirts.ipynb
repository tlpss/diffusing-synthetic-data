```python
tshirt_scenes_path = "/home/tlips/Documents/diffusing-synthetic-data/data/scenes/tshirts/syncloth-filtered-2500-ordered"

import pathlib 

tshirt_scenes_path = pathlib.Path(tshirt_scenes_path)
```

```python
json_files = list(tshirt_scenes_path.glob("**/*.json"))
len(json_files)

```

```python
from dsd.rendering.order_tshirt_keypoints import order_tshirt_keypoints
import json
import numpy as np 
```

```python
for json_file in json_files:
    kp_dict = json.load(open(json_file))
    # print(kp_dict)
    values = kp_dict.values()
    x_min = min([v[0] for v in values])
    x_max = max([v[0] for v in values])
    y_min = min([v[1] for v in values])
    y_max = max([v[1] for v in values])
    bbox = [x_min,y_min,x_max-x_min,y_max-y_min]
    kp_values = list(kp_dict.values())
    kp_values = np.array(kp_values)

    kp_values = order_tshirt_keypoints(kp_values, bbox)

    kps = kp_values.tolist()
    kps = [kps[i] for i in range(len(kps))]
    kp_dict = {k: v for k,v in zip(kp_dict.keys(), kps)}

    # print(kp_dict)
    # print("--")

    with open(json_file, "w") as f:
        json.dump(kp_dict, f)
```
